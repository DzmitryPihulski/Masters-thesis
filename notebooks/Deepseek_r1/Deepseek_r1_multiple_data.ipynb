```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_1 = pd.read_parquet('data/data_with_nationality_1.parquet')
data_2 = pd.read_parquet('data/data_with_nationality_2.parquet')
data_openai = pd.read_parquet('data/data_with_nationality_1_openai.parquet')
```

```python
data_2 = data_2.iloc[:214, :]
data_1 = data_1.iloc[:214, :]
data_openai = data_openai.iloc[:214, :]
```

```python
for colname in [col for col in data_1.columns if col.endswith('_answer')]:
    print(sum(data_1[colname] != data_2[colname]))
```

```python
for colname in [col for col in data_1.columns if col.endswith('_answer')]:
    print(sum(data_1[colname] != data_openai["OPENAI" + colname[8:]]))
```

```python
first = data_1[[col for col in data_1.columns if col.endswith('_answer')]]
second = data_2[[col for col in data_1.columns if col.endswith('_answer')]]

cross_corr = pd.DataFrame(
    index=first.columns,
    columns=second.columns
)

# Fill the correlation matrix
for col1 in first.columns:
    for col2 in second.columns:
        cross_corr.loc[col1, col2] = first.corr(second[col2])

print("Cross-dataframe correlation matrix:")
print(cross_corr)

# Visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(cross_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('Correlation Between Columns of Different Dataframes')
plt.tight_layout()
plt.show()
```
