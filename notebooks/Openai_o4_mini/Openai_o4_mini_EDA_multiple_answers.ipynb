```python
import pandas as pd
import numpy as np
```

```python
data_1 = pd.read_parquet('data/Openai/data_with_nationality_1_openai.parquet', engine='pyarrow')
data_2 = pd.read_parquet('data/Openai/data_with_nationality_2_openai.parquet', engine='pyarrow')
data_3 = pd.read_parquet('data/Openai/data_with_nationality_3_openai.parquet', engine='pyarrow')
data_4 = pd.read_parquet('data/Openai/data_with_nationality_4_openai.parquet', engine='pyarrow')
data_5 = pd.read_parquet('data/Openai/data_with_nationality_5_openai.parquet', engine='pyarrow')
```

```python
new_data = data_1.copy()
```

```python
new_data = new_data.drop(new_data.columns.difference(["Text_EN", "Text_PL", "Text_RU", "Translation_quality"]), axis=1)
```

```python
target_columns = ['far_right_EN', 'far_right_PL', 'far_right_RU', 'mod_cons_EN', 'mod_cons_PL', 'mod_cons_RU',
       'prog_left_EN', 'prog_left_PL', 'prog_left_RU', 'centrist_EN', 'centrist_PL', 'centrist_RU']
for target_column in target_columns:
    new_data[f'OPENAI_{target_column}_answer'] = None
new_data.head()
```

```python
colnames = [col for col in new_data.columns if col.endswith('_answer')]
# For each column that ends with "_answer"
valid = 0
for index, row in new_data.iterrows():
    for i in range(len(colnames)):
        if row['Translation_quality'] == 1.0:
            column_sum = data_1.loc[index, colnames[i]] + data_2.loc[index, colnames[i]] + data_3.loc[index, colnames[i]] + data_4.loc[index, colnames[i]] + data_5.loc[index, colnames[i]]
            
            if (column_sum >=4 or column_sum <= 1):
                valid += 1
                if column_sum >= 4:
                    new_data.loc[index, colnames[i]] = 1
                else:
                    new_data.loc[index, colnames[i]] = 0

print(valid, valid/(new_data[new_data['Translation_quality'] == 1.0].shape[0]*12), (new_data[new_data['Translation_quality'] == 1.0].shape[0]*12) - valid)
```
